## Data Visualization

```python
region_sales_value = df.groupby("Region")["Sales_Volume"].sum().reset_index()
plt.pie(region_sales_value["Sales_Volume"],labels=region_sales_value["Region"],autopct="%1.1f%%")
plt.title("Pie Chart of Region Sales")
plt.figure(figsize=(10,6))
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(),cmap="seismic",annot=True)
plt.title("Correlation of all numerical values")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x="Model",y="Sales_Volume",palette="viridis")
plt.title("All models Sales",fontsize = 18)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df, x="Year", y="Sales_Volume", marker="o")
plt.show()
```

## Feature Engineering

```python
df.head(1)
```

```python
from sklearn.preprocessing import OneHotEncoder # Create OneHotEncoder (used to convert categorical values into numerical form)
```

```python
encoder = OneHotEncoder()

```

```python
cat_values = df[["Model","Region","Color","Fuel_Type","Transmission"]] # This all are categoricals
```

```python
cat_to_numerical = encoder.fit_transform(cat_values).toarray()
# Fit the encoder on categorical values and transform them into numerical (one-hot encoded) array
```

```python
cat_to_numerical
```

```python
convert_to_df = pd.DataFrame(cat_to_numerical,columns=encoder.get_feature_names_out())
# Convert the encoded numerical array into a DataFrame with proper column names
```

```python
convert_to_df.head()
```

```python
final_df = pd.concat([convert_to_df,df],axis=1) # Combine the original DataFrame with the encoded DataFrame (side by side)
```

```python
final_df
```

```python
final_df.info()
```

```python
final_df.Sales_Classification.unique()
```

```python
final_df["Sales_Classification"] = final_df["Sales_Classification"].map({'High':1, 'Low':0})
# Convert categorical column 'Sales_Classification' into numerical values (High = 1, Low = 0)
```

```python
# Drop unnecessary original categorical columns (since we already encoded them)
final_df.drop(labels="Color",axis=1,inplace=True)
final_df.drop(labels="Transmission",axis=1,inplace=True)
final_df.drop(labels="Fuel_Type",axis=1,inplace=True)
final_df.drop(labels="Region",axis=1,inplace=True)
final_df.drop(labels="Model",axis=1,inplace=True)
```
