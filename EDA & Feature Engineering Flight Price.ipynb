```python
df["Day"] = df["Date_of_Journey"].str.split("/").str[0]
df["Month"] = df["Date_of_Journey"].str.split("/").str[1] # extracting Day, Month, and Year from a column Date_of_Journey
df["Year"] = df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head(1)
```

```python
df.drop(labels="Date_of_Journey",axis=1,inplace=True)
```

```python
df["Arrival_Time"] = df["Arrival_Time"].str.split(" ").str[0]
```

```python
df.head(1)
```

```python
df["Arrival_Hours"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_Minutes"] = df["Arrival_Time"].str.split(":").str[1] # splitting the Arrival_Time column into hours and minutes
```

```python
df.drop(labels="Arrival_Time",axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df.Additional_Info.unique() # remove the columns if not effected price 
```

```python
df.drop(labels="Additional_Info",axis=1,inplace=True)
```

```python
df["Dep_Hours"] = df["Dep_Time"].str.split(":").str[0]
df["Dep_Minutes"] = df["Dep_Time"].str.split(":").str[1]
```

```python
df.drop(labels="Dep_Time",axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df["Duration_Hours"] = df["Duration"].str.split(" ").str[0].str.split("h").str[0]
df["Duration_Minutes"] = df["Duration"].str.split(" ").str[1].str.split("m").str[0]
```

```python
df.head(1)
```

```python
df.drop(labels="Duration",axis=1,inplace=True)
```

```python
df.head(1)
```

```python
df.Total_Stops.unique()
```

```python
df["Total_Stops"] = df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4})
# converting the categorical column Total_Stops into numerical values
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.Duration_Hours.unique()
```

```python
df = df[df["Duration_Hours"].str.isnumeric()] # If we just want to drop the row with "5m", it convert to numerical value
```

```python
df.Duration_Hours.unique()
```

```python
# converting all those columns to nullable integers (Int64 in pandas)

df["Day"] = df["Day"].astype("Int64")
df["Month"] = df["Month"].astype("Int64")
df["Year"] = df["Year"].astype("Int64")
df["Arrival_Hours"] = df["Arrival_Hours"].astype("Int64")
df["Arrival_Minutes"] = df["Arrival_Minutes"].astype("Int64")
df["Dep_Hours"] = df["Dep_Hours"].astype("Int64")
df["Dep_Minutes"] = df["Dep_Minutes"].astype("Int64")
df["Duration_Hours"] = df["Duration_Hours"].astype("Int64")
df["Duration_Minutes"] = df["Duration_Minutes"].astype("Int64")
```

```python
cat_values = df[["Airline","Source","Destination","Route"]] 
```

```python
encoder = OneHotEncoder()
```

```python
cat_to_numerical = encoder.fit_transform(cat_values).toarray() # convert to categorical to numerical values 
```

```python
cat_to_numerical
```

```python
df1 = pd.DataFrame(cat_to_numerical,columns=encoder.get_feature_names_out())
```

```python
df1.head(4)
```

```python
df.drop(labels="Airline",axis=1,inplace=True)
df.drop(labels="Source",axis=1,inplace=True)
df.drop(labels="Destination",axis=1,inplace=True)
df.drop(labels="Route",axis=1,inplace=True)
```

```python
concat_two_dataframe = pd.concat([df,df1],axis=1)
```
